```python
X = df.iloc[:, 0:4]
y = df.iloc[:, 4]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)

print(X_train.head())
print('*'*12)
print(X_test.head())
```

```python
from sklearn.preprocessing import LabelEncoder
e = LabelEncoder()
e.fit(y_train)
y_train = e.transform(y_train)

print('y_train: ', y_train)
```

```python
y_train_transformed = e.transform(['setosa', 'versicolor', 'virginica'])
print('setosa:{0}, versicolor:{1}, virginica:{2}'.format(y_train_transformed[0], y_train_transformed[1], y_train_transformed[2]))
```

```python
import tensorflow as tf
y_train = tf.keras.utils.to_categorical(y_train)
print(y_train)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(16, input_dim=4, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])
model.fit(X_train, y_train, epochs=50, batch_size=1)
```

```python
y_test = e.transform(y_test)

print(y_test)
```

```python
import tensorflow as tf
y_test = tf.keras.utils.to_categorical(y_test)
print(y_test)
```

```python
print('Test Data Accuracy: ', model.evaluate(X_test, y_test)[1])
```

```python
import pandas as pd

df = pd.read_csv('C:/pytest/wine.csv', header=None)
df = df.sample(frac=1)

X = df.iloc[:, :12]
y = df.iloc[:, 12]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = Sequential()
model.add(Dense(30, activation='relu', input_dim=12))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
```

```python
history = model.fit(X_train, y_train, epochs=200, batch_size=200, validation_split=0.2)
```

```python
acc = list(map(lambda x:round(x,3),history.history['acc']))
val_acc = list(map(lambda x:round(x,3),history.history['val_acc']))
loss = history.history['loss']
val_loss = list(map(lambda x:round(x,3),history.history['val_loss']))

print('Validation accuracy of each epoch:', val_acc)
```

```python
type(history.history['val_loss'])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
epochs = range(1, len(acc)+1)
plt.figure(figsize=(30,20))
plt.plot(epochs, acc, 'bo', label='Training Acc')
plt.plot(epochs, val_acc, 'b', label='Validation Acc')
plt.title('Training and validation accuracy')
plt.legend()
plt.show()
```

```python
max_val = max(val_acc)
print(max_val, val_acc.index(max_val))
```
