```python
bank_list = []

for idx in lines:
    if '금융' in idx:
        temp = idx.rstrip('\n')
        bank_list.append(temp)

bank_df = pd.DataFrame(bank_list, columns = ['line'])
```

# 연습문제 5

* 연습 폴더에 있는 ‘금융규제운영규정.txt’ 파일을 읽고 사용자가 입력한 키워드가 있는 문장을 찾고, 찾은 키워드를 보여주는 코드를 작성하시오
* 찾는 키워드는 먼저 하나만 찾게 하고, 가능하면 복수 개가 가능하도록 하세요
* 분석 결과는 다음과 같이 나오게 하며, 가능하면 저장 기능까지 구현하세요
* 형태소 분석기는 사용하지 않아도 됨

```python
import os
import pandas as pd
path = 'C:/pytest/'

f = open(path + '금융규제운영규정.txt', 'r')
lines = f.readlines()

org_lines = []

for line in lines:
    temp = line.rstrip('\n')
    org_lines.append(temp)
```

```python
kwd = input()
kwd_list = []

for line in org_lines:
    if kwd in line:
        kwd_list.append(line)

often_list = []

for line in kwd_list:
    temp = line.count(kwd)
    temp_kwd = (kwd + ',')*temp
    often_list.append(temp_kwd)

kwd_df = pd.DataFrame({kwd : kwd_list, kwd+' 빈도' : often_list})
kwd_df.to_csv(path + kwd+'_inline_input.csv', index = False, encoding= 'CP949')
```

```python
kwd = list(input().split())
kwd_list = []

for key in kwd:
    temp = []
    for line in org_lines:
        if key in line:
            temp.append(line)
    kwd_list.append(temp)

often_list = []

for key, line in zip(kwd, kwd_list):
    temp_list = []
    for l in line:
        temp = l.count(key)
        temp_kwd = (key + ',')*temp
        temp_list.append(temp_kwd)
    often_list.append(temp_list)

df_list = []

for k, key, often in zip(kwd, kwd_list, often_list):
    temp = pd.DataFrame({k : key, k+' 빈도' : often})
    df_list.append(temp)

for key, df in zip(kwd, df_list):
    df.to_csv(path + key+'_inline_multi_input.csv', index = False, encoding= 'CP949')
```
